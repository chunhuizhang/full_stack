## docker vs. docker-desktop

- 先安装 docker 再安装 docker-desktop
    - 安装 docker-desktop 会创建以及依赖认证和校验之类：`~/.docker/config.json`
    - 会导致之前的 docker images/docker containers 失效（无法通过 docker images、docker ps 查询）；

- `docker version -f "{{.Server.Os}}"`: `linux`/`windows`
    - `-f`: `--format`
    - client/server
        - Client: Docker Engine - Community
        - Server: Docker Desktop 4.19.0 (12345)
    - 如果返回 "linux"，则 Docker 服务器正在运行 Linux 容器。如果返回 "windows"，则 Docker 服务器正在运行 Windows 容器。
