```python
from graphviz import Digraph
from IPython.display import display
```

```python
dot = Digraph(comment='Minimax Tree')

# 为不同玩家设置不同颜色的节点
dot.node('A', 'Root', color='lightblue2', style='filled')
dot.node('B', 'Max 1', color='lightblue2', style='filled')
dot.node('C', 'Min 1', color='lightcoral', style='filled')
dot.node('D', 'Max 2', color='lightblue2', style='filled')
dot.node('E', 'Max 3', color='lightblue2', style='filled')

dot.edges(['AB', 'AC'])
dot.edge('B', 'D')
dot.edge('B', 'E')

# 直接在Jupyter Notebook中显示图像
display(dot)
```

## 属性

- dot.attr
    - rankdir
        - TB
        - BT
- color
    - `white`/`black`
    - `lightblue2`/`lightcoral`
- style
    - `filled`
