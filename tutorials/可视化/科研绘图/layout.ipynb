```python
import matplotlib.pyplot as plt
import numpy as np

figure_mosaic = """
ABCC
DDCC
DDEF
"""

fig, ax = plt.subplot_mosaic(mosaic=figure_mosaic, figsize=(10, 5))
fig.suptitle('Easy subplotting with subplot_mosaic')

x = np.linspace(0.1, 100, 1000)
ax['A'].plot(x, np.sin(x), 'r-')
ax['B'].plot(x, np.cos(x), 'g-')
ax['C'].plot(x, np.sin(x)*np.cos(x), 'b-')

ax['D'].plot(x, x**2, 'r-')
ax['E'].plot(x, np.sqrt(x), 'g-')
ax['F'].plot(x, 1/np.sqrt(x), 'b-')

plt.tight_layout()
plt.savefig('./figs/subplot_mosaic.svg', format='svg')
```
