- style
    - `sns.set_style("whitegrid")`
- `histplot` vs. `countplot`
    - `histplot` 数值（某列的元素为数值）直方图；
    - `countplot` 类别型（某列的元素为 `string`/`categorical` ）直方图；
- scatterplot
    - `ax = sns.scatterplot(x='', y='', hue='', data=df)`
        - x, y, hue 都是 df 的列名；
- regplot
    - logistics=True
