## cmap

- cmap
    - `'RdYlGn'`

```python
def f(x, y):
    return x**2 + y**2
```

```python
# xs, ys = np.meshgrid(np.linspace(-1, 1, 100), np.linspace(-1, 1, 100))
# in_circle = (f(xs, ys) <= 1).astype(int)

xs = np.random.uniform(low=-1, high=1, size=(10000, ))
ys = np.random.uniform(low=-1, high=1, size=(10000, ))
in_circle = (f(xs, ys) <= 1).astype(int)
```

```python
plt.scatter(xs, ys, c=in_circle, cmap='RdYlGn', vmin=-0.1, vmax=1.1, lw=0, s=2)
plt.axis('equal')
```
