## runtime vs. compile time

- compile time：编译期，runtime：运行时
    - 编译不通过，编译期错误；
        - 编译期错误相对好排查，运行时错误需要到真正运行时才好排查（甚至可能断点跟才能发现）
    - 一个代码程序需要经历的两个阶段；
        - 若果有错误，越早发现越好，也即尽量在编译期发现；
        - 中间还会有一个阶段：link-time，链接期；

- 运行时
    - 运行时库：runtime library，这些还都是代码层面
    - 运行时环境：runtime environment


```
// C 语言的标准库
// windows CRT: C runtime library
// linux: glibc GNU C library
#include <stdio.h>

```

- 运行时环境
    - 无运行时：（直接）依赖于操作系统，而 os 也可以视为一种 runtime
        - C/C++
        - Rust
        
    - 轻运行时
        - golang：运行时和代码进行打包在一起；
        
    - 重运行时：运行时需要单独安装；
        - Java（JVM）
        - Python（CPython）
        - C#（.Net runtime）
